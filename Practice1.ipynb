## Краткая справка по функции input() в Python
Функция input() используется для получения ввода от пользователя через консоль. Она приостанавливает выполнение программы и ждет, пока пользователь введет данные и нажмет Enter.

Синтаксис:

```python
input([prompt])
```

prompt (необязательный аргумент): строка, которая выводится на экран перед ожиданием ввода. Это может быть подсказка для пользователя.

Возвращаемое значение:
Функция всегда возвращает введенные данные в виде строки (тип str).

Примеры использования:

### 1. Простой ввод:

```python
name = input("Введите ваше имя: ")
print(f"Привет, {name}!")
```

### 2.Преобразование ввода:
Поскольку input() возвращает строку, для работы с числами нужно преобразовать ввод в нужный тип (например, int или float).

```python
age = int(input("Введите ваш возраст: "))
print(f"Через 5 лет вам будет {age + 5} лет.")
```

### 3.Ввод нескольких значений:
Можно использовать split() для разделения ввода на несколько значений.

```python
x, y = input("Введите два числа через пробел: ").split()
x = int(x)
y = int(y)
print(f"Сумма: {x + y}")
```

### 4.Обработка ошибок:
Если пользователь вводит некорректные данные (например, буквы вместо чисел), можно использовать обработку исключений.

```python
try:
    number = int(input("Введите число: "))
    print(f"Вы ввели: {number}")
except ValueError:
    print("Ошибка: введено не число!")
```

Важные моменты:
Ввод всегда возвращается как строка, даже если пользователь вводит число.

Если нужно работать с числами, используйте функции преобразования: int(), float() и т.д.

Для ввода нескольких значений можно использовать split() или несколько вызовов input().

Функция input() — это простой и удобный способ взаимодействия с пользователем в консольных приложениях. При решении задач ниже вы должны использовать эту функцию, так что можете обращаться к данной справке за помощью.

## Задача 4
***
Напишите программу, которая симулирует движение робота по квадратной траектории. Робот должен двигаться вперед, поворачивать на 90 градусов и повторять это 4 раза.

```python
import time

print("Выберите длину стороны квадрата (м): ")
side_lenght = float(input())
print("Выберите скорость движения робота (м/с): ")
speed = float(input())
delay = 1/speed

for i in range(4):
  print("Сторона: ", i+1)
  print("Движение на", side_lenght, "метров")
  time.sleep(delay)

  if (i<3):
    print("Поворот на 90 градусов")
    time.sleep(delay/2)

print("Пройденное расстояние:", side_lenght*4, "метров")
```

```python

power = 100
print("Изначальный заряд робота:", power,"%")
while (power != 0) :
  print("Выберите действие (1(10%),2(25%),3(5%),4(37%)):")
  x = float(input())
  if x > 4 or x < 1:
     print("Ошибка, введите действительное действие (1,2,3,4):")
     x = float(input())

  match x:
    case 1:
      if (power >= 10):
        power = power - 10
        print("Текущий заряд:",power,"%")
      else:
         print("Текущий заряд:",power,"%")
         print("недостаточно заряда для действия:",x)
    case 2:
      if (power >= 25):
        power = power - 25
        print("Текущий заряд:",power,"%")
      else:
        print("Текущий заряд:",power,"%")
        print("недостаточно заряда для действия:",x)
    case 3:
      if (power >= 5):
        power = power - 5
        print("Текущий заряд:",power,"%")
      else:
        print("Текущий заряд:",power,"%")
        print("недостаточно заряда для действия:",x)
    case 4:
      if (power >= 37):
        power = power - 37
        print("Текущий заряд:",power,"%")
      else:
        print("Текущий заряд:",power,"%")
        print("недостаточно заряда для действия:",x)

  if (power == 0):
    print("Робот разрядился")
  if power < 5:
    print("недостаточно заряда для любого действия")
    break
```
